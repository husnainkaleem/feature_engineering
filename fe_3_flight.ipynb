**Objective:**

The objective of this project is to develop a machine learning model that predicts flight prices accurately. By analyzing historical flight data and relevant features, the model aims to forecast the prices of future flights with minimal error. The predicted prices can assist travelers in planning their trips more effectively and help airlines optimize pricing strategies to improve revenue management.

**Dataset Description:**

The dataset contains information about past flight itineraries, including attributes such as departure time, arrival time, total stops, destination, source, and ticket prices.

**Feature engineering**

This code below is extracting the day, month, and year components from the 'Date_of_Journey' column in the DataFrame final_df and storing them in separate columns 'Date', 'Month', and 'Year', respectively.

```python
##Fetaure Engineering Process
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head(2)
```

**the code below convert the "date","month" and "year" column in integer data type**

```python
final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.info()
```

**dropping the "date_of_journey" column**

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head(10)
```

**the code below  extracts the arrival time component from each timestamp in the 'Arrival_Time' column.**

```python
final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
```

**This code below is extracting the hour and minute components from the 'Arrival_time' column in the DataFrame final_df and storing them in separate columns 'Arrival_hour' and 'Arrival_min' respectively.**

```python
final_df['Arrival_hour']=final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_min']=final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df.head(1)
```

**the code below convert the variable in integer data type**

```python
final_df['Arrival_hour']=final_df['Arrival_hour'].astype(int)
final_df['Arrival_min']=final_df['Arrival_min'].astype(int)
```

**dropping arrival time column**

```python
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df.head(5)
```

**This code below is extracting the hour and minute components from the 'Dep_time' column in the DataFrame final_df and storing them in separate columns 'Dept_hour' and 'Dep_min' respectively. and convert the variable in the integer data type.**

---



---
